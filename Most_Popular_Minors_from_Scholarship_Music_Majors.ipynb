```python
all_registration = pd.read_csv("~/Downloads/UsdanRegistration2016-2018.csv")

all_registration.dropna(axis=0, subset=['Major'])
all_registration
```

```python
eighteen = all_registration.query('year==2018')
eighteen
```

```python
scholarships = eighteen.dropna(axis=0, subset=['_ScholarshipTotal'])
scholarships = eighteen.query('_TuitionAidTotal != 0.0 or _ScholarshipTotal != 0.0')
# eighteen.dropna(axis=0, subset=['_TuitionAidTotal']).query('_TuitionAidTotal != 0.0')
# scholarshiphs = scholarships.dropna(axis=0, subset=['_ScholarshipTotal']).query('_ScholarshipTotal !=0.0')
scholarships
```

```python
music_scholarships = scholarships.query('dept_t == "Orchestra" or dept_t == "Chorus" or dept_t == "Band"')
music_scholarships
```

```python
non_music_minors = non_music_minors = music_scholarships.query('Minor2 != "Absolute Beginning Guitar" and Minor2 != "Absolute Beginner Piano" and Minor2 != "Chamber Music" and Minor2 != "Guitar & Songwriting" and Minor2 != "Jazz Ensemble" and Minor2 != "Junior Chorus" and Minor2 != "Piano" and Minor2 != "Rock On!" and Minor2 != "Senior Chorus"')
non_music_minors = non_music_minors.sort_values('Minor2', ascending = False)
only_minor2 = non_music_minors['Minor2'].value_counts()
only_minor2
```
